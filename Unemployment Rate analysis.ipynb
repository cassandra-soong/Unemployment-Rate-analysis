### Unemployment rate in U.S and U.K between 1991 - 2021

```python
uk_us = df.loc[["United States", "United Kingdom"], :]
uk_us = uk_us.drop(["Country Code"], axis=1).T
uk_us.head()
```

```python
fig, ax = plt.subplots(figsize=(10,4))

ax.plot(
    uk_us.index,
    uk_us["United States"],
    color = "red",
    linewidth = 2,
    label = "United States"
)

ax.plot(
    uk_us.index,
    uk_us["United Kingdom"],
    color = "blue",
    linewidth = 2,
    label = "United Kingdom"
)

ax.set_title("Unemployment rate in U.S and U.K between 1991-2021")
ax.set_xlabel("Year")
ax.set_ylabel("Unemployment rate (%)")
ax.set_xticks(uk_us.index[::2])
ax.axvline(16.5, c="black", ls="--")
ax.text(9, 7, "Global Financial Crisis-->")
ax.legend()

plt.show()
```

#### The unemployment rate in U.S and U.K both had similar trends. It decreased drastically over the years between 1993-2001 where more job opportunities were available. However, the unemployment rate spiked again between year 2008-2009 due to the Global Financial Crisis.
